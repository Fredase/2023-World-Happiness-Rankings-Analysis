```python
import random

# Define the game map
map = {
    "Location 1": "Description of location 1",
    "Location 2": "Description of location 2",
    "Location 3": "Description of location 3",
    "Location 4": "Description of location 4",
    "Location 5": "Description of location 5",
}

# Define the game characters
characters = {
    "Character 1": "Description of character 1",
    "Character 2": "Description of character 2",
}

# Define the game items
items = {
    "Item 1": "Description of item 1",
    "Item 2": "Description of item 2",
}

# Define the player's starting health
health = 100

# Define the player's inventory and wallet
inventory = []
wallet = 0

# Define the game loop
while True:
    # Print the current location and its description
    current_location = random.choice(list(map.keys()))
    print("You are at", current_location)
    print(map[current_location])

    # Print the informational greeting for the location
    print("Welcome to", current_location)

    # Prompt the player for an action
    action = input("What would you like to do? ")

    # Process the player's action
    if action == "talk":
        # Select a random character to talk to
        current_character = random.choice(list(characters.keys()))
        print("You talk to", current_character)
        print(characters[current_character])
    elif action == "take":
        # Select a random item to take
        current_item = random.choice(list(items.keys()))
        print("You take", current_item)
        inventory.append(current_item)
    elif action == "use":
        # Prompt the player to select an item from their inventory
        if len(inventory) == 0:
            print("Your inventory is empty")
        else:
            print("Select an item to use:")
            for i, item in enumerate(inventory):
                print(i+1, item)
            item_index = int(input()) - 1
            if item_index < 0 or item_index >= len(inventory):
                print("Invalid item selection")
            else:
                # Use the selected item
                used_item = inventory[item_index]
                print("You use", used_item)
                inventory.remove(used_item)
    elif action == "buy":
        # Prompt the player to enter the amount of money they want to spend
        money_spent = int(input("How much money do you want to spend? "))
        if money_spent > wallet:
            print("You don't have enough money")
        else:
            # Subtract the money spent from the wallet
            wallet -= money_spent
            print("You spent", money_spent, "coins")
    elif action == "quit":
        # End the game loop
        break
    else:
        print("Invalid action")

    # Decrease the player's health by a random amount
    health -= random.randint(1, 10)
    if health <= 0:
        # End the game if the player's health reaches zero
        print("You died")
        break
```

```python
import random

# Define the locations in the game
locations = {
    "home": "You are in your spaceship, ready to embark on a journey through space.",
    "planet1": "You have landed on a desolate planet with red sand and rocky terrain.",
    "planet2": "You have arrived on a lush green planet with strange alien creatures.",
    "planet3": "You have discovered a mysterious planet with a hidden civilization.",
    "wormhole": "You have found a wormhole that could take you to another part of the universe."
}

# Define the characters in the game
characters = {
    "dragon": "A bored space dragon appears and asks if you need any help.",
    "alien": "A friendly alien approaches you and offers to trade goods."
}

# Define the player's health and inventory
health = 100
inventory = []

# Define the player's wallet
wallet = 0

# Define a function to print the player's status
def print_status():
    print("Health:", health)
    print("Inventory:", inventory)
    print("Wallet:", wallet)

# Define a function to handle interactions with characters
def interact(character):
    print(characters[character])
    response = input("What would you like to do? ")
    if response == "trade":
        trade()
    elif response == "help":
        help()
    else:
        print("I don't understand.")

# Define a function for trading with the alien
def trade():
    global wallet
    item = random.choice(["laser gun", "food ration", "med kit"])
    price = random.randint(1, 10)
    print("The alien offers you a", item, "for", price, "credits.")
    response = input("Do you want to buy it? ")
    if response == "yes" and wallet >= price:
        print("You bought the", item)
        wallet -= price
        inventory.append(item)
    elif response == "yes" and wallet < price:
        print("You don't have enough credits.")
    else:
        print("You didn't buy anything.")

# Define a function for getting help from the dragon
def help():
    global health
    print("The dragon gives you a health boost.")
    health += 10

# Define the main game loop
while True:
    print_status()
    location = input("Where would you like to go? ")
    if location in locations:
        print(locations[location])
        input("Press enter to continue.")
        if location == "home":
            interact("dragon")
        elif location == "planet1":
            interact("alien")
        elif location == "planet2":
            interact("alien")
        elif location == "planet3":
            interact("dragon")
        else:
            print("There is nothing to do here.")
    elif location == "wormhole":
        print("You enter the wormhole and find yourself in a new part of the universe.")
    else:
        print("That is not a valid location.")
    
    # Check if the player's health is zero or less
    if health <= 0:
        print("You died.")
        break
        
    # Check if the player has reached the end of the game
    if location == "end":
        print("You have reached the end of the game!")
        break
```
